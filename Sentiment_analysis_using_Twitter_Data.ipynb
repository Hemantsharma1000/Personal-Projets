```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('/content/twitter_training.csv')
```

```python
train_data = train_data.dropna()
```

```python
X_train = train_data['Text']
y_train = train_data['Sentiment']
```

```python
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
```

```python
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_train_vec)
accuracy = accuracy_score(y_train, y_pred)
print("Accuracy:", accuracy)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment', data=train_data, palette="Set2")
plt.title("Sentiment Distribution in Training Data")
plt.show()
```

```python
new_data = pd.read_csv('/content/twitter_validation.csv')
```

```python
new_data = new_data.dropna()
X_new = new_data['text']
```

```python
X_new_vec = vectorizer.transform(X_new)
```

```python
y_pred_new = model.predict(X_new_vec)
new_data['predicted_sentiment'] = y_pred_new
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='predicted_sentiment', data=new_data, palette="Set2")
plt.title("Predcted Sentiment Distribution in New Data")
plt.show()
```
